```python
import random
import time
import matplotlib.pyplot as plt

def heapify(arr, n, i):
    largest = i 
    l = 2 * i + 1
    r = 2 * i + 2
    
    if l < n and arr[l] > arr[largest]:
        largest = l
    if r < n and arr[r] > arr[largest]:
        largest = r
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heapSort(arr):
    n = len(arr)
    for i in range(n//2 - 1, -1, -1):
        heapify(arr, n, i)
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

sizes = [1000, 5000, 10000, 25000, 50000, 100000]
heap_times = []

for N in sizes:
    data = [random.randint(1, 10000) for _ in range(N)]
    
    arr = data[:]
    start = time.time()
    heapSort(arr)
    heap_times.append(time.time() - start)

plt.plot(sizes, heap_times, label="Heap Sort", marker="o")
plt.xlabel("Input Size (N)")
plt.ylabel("Time (seconds)")
plt.title("Heap Sort Performance")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt
def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        mergeSort(L)
        mergeSort(R)

        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

sizes = [1000, 5000, 10000, 25000, 50000, 100000]
merge_times = []

for N in sizes:
    data = [random.randint(1, 10000) for _ in range(N)]
    
    arr = data[:]
    start = time.time()
    mergeSort(arr)
    merge_times.append(time.time() - start)

plt.plot(sizes, merge_times, label="Merge Sort", marker="o", color="orange")
plt.xlabel("Input Size (N)")
plt.ylabel("Time (seconds)")
plt.title("Merge Sort Performance")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt


def partition(arr, low, high):
    i = low - 1
    pivot = arr[high]  

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i + 1

def quickSort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quickSort(arr, low, pi - 1)
        quickSort(arr, pi + 1, high)


sizes = [1000, 5000, 10000, 25000, 50000, 100000]
quick_times = []

for N in sizes:
    data = [random.randint(1, 10000) for _ in range(N)]
    
    arr = data[:]
    start = time.time()
    quickSort(arr, 0, len(arr) - 1)
    quick_times.append(time.time() - start)


plt.plot(sizes, quick_times, label="Quick Sort", marker="o", color="green")
plt.xlabel("Input Size (N)")
plt.ylabel("Time (seconds)")
plt.title("Quick Sort Performance")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt

def selectionSort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
sizes = [1000, 2000,3000,5000,10000]  
selection_times = []

for N in sizes:
    data = [random.randint(1, 10000) for _ in range(N)]
    
    arr = data[:]
    start = time.time()
    selectionSort(arr)
    selection_times.append(time.time() - start)

plt.plot(sizes, selection_times, label="Selection Sort", marker="o", color="red")
plt.xlabel("Input Size (N)")
plt.ylabel("Time (seconds)")
plt.title("Selection Sort Performance")
plt.legend()
plt.grid(True)
plt.show()   
```

```python
import random
import time
import matplotlib.pyplot as plt 
%matplotlib inline

def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
       
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

sizes = [500, 1000, 2000, 3000, 4000]  
insertion_times = []

for N in sizes:
    data = [random.randint(1, 10000) for _ in range(N)]
    
    arr = data[:]
    start = time.time()
    insertionSort(arr)
    insertion_times.append(time.time() - start)

plt.plot(sizes, insertion_times, label="Insertion Sort", marker="o", color="blue")
plt.xlabel("Input Size (N)")
plt.ylabel("Time (seconds)")
plt.title("Insertion Sort Performance")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
import time
import matplotlib.pyplot as plt

sizes_fast = [1000, 5000, 10000, 25000, 50000, 100000]   
sizes_slow = [500, 1000, 2000, 3000, 4000]              

heap_times, merge_times, quick_times = [], [], []
selection_times, insertion_times = [], []

for N in sizes_fast:
    data = [random.randint(1, 10000) for _ in range(N)]

    arr = data[:]
    start = time.time()
    heapSort(arr)
    heap_times.append(time.time() - start)

    arr = data[:]
    start = time.time()
    mergeSort(arr)
    merge_times.append(time.time() - start)

    arr = data[:]
    start = time.time()
    quickSort(arr, 0, len(arr)-1)
    quick_times.append(time.time() - start)

for N in sizes_slow:
    data = [random.randint(1, 10000) for _ in range(N)]

    arr = data[:]
    start = time.time()
    selectionSort(arr)
    selection_times.append(time.time() - start)

    arr = data[:]
    start = time.time()
    insertionSort(arr)
    insertion_times.append(time.time() - start)

plt.figure(figsize=(10,6))

plt.plot(sizes_fast, heap_times, label="Heap Sort", marker="o")
plt.plot(sizes_fast, merge_times, label="Merge Sort", marker="o")
plt.plot(sizes_fast, quick_times, label="Quick Sort", marker="o")
plt.plot(sizes_slow, selection_times, label="Selection Sort", marker="o")
plt.plot(sizes_slow, insertion_times, label="Insertion Sort", marker="o")

plt.xlabel("Input Size (N)")
plt.ylabel("Time (seconds)")
plt.title("Comparison of Sorting Algorithms")
plt.legend()
plt.grid(True)
plt.show()
```
